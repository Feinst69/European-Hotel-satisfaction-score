```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the data
file_path = 'C:/Users/tanjo/Documents/Cours/SQL/Europe Hotel Booking Satisfaction Score.csv'
data = pd.read_csv(file_path)

# Select features and target variable
rating_columns = ['Hotel wifi service', 'Departure/Arrival  convenience', 'Ease of Online booking', 'Hotel location', 'Food and drink', 'Stay comfort', 'Common Room entertainment', 'Checkin/Checkout service', 'Other service', 'Cleanliness']
target_column = 'satisfaction'

# Encode categorical variables
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    if column != target_column:
        label_encoders[column] = LabelEncoder()
        data[column] = label_encoders[column].fit_transform(data[column])

# Encode the target variable
target_encoder = LabelEncoder()
data[target_column] = target_encoder.fit_transform(data[target_column])

# Handle missing values (if any)
data = data.dropna()

# Split the data into features and target
X = data[rating_columns]
y = data[target_column]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create age bins
age_bins = [0, 18, 30, 45, 60, 100]
age_labels = ['0-18', '19-30', '31-45', '46-60', '61+']
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)

# Define the services to plot
services = ['Common Room entertainment', 'Checkin/Checkout service', 'Other service', 'Ease of Online booking',  'Hotel location', 'Food and drink',  'Stay comfort']

# Ensure the service columns are numeric
for service in services:
    data[service] = pd.to_numeric(data[service], errors='coerce')

# Calculate the average ratings for each service across age bins
average_ratings = data.groupby('Age Group')[services].mean().reset_index()

# Melt the DataFrame for easier plotting with seaborn
melted_data = average_ratings.melt(id_vars='Age Group', var_name='Service', value_name='Average Rating')

# Plotting the results using seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Age Group', y='Average Rating', hue='Service', data=melted_data)
plt.title('Average Service Ratings by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Rating')
plt.legend(title='Service')
plt.xticks(rotation=45)
plt.savefig('Rate of the services from each age bins.png')

# Identify and print the two lowest-rated services for each age bin
print("Two lowest-rated services for each age bin:")
for age_group in age_labels:
    avg_ratings_for_age_group = average_ratings[average_ratings['Age Group'] == age_group].set_index('Age Group')
    lowest_two_services = avg_ratings_for_age_group.T.nsmallest(2, age_group).index.tolist()
    lowest_two_values = avg_ratings_for_age_group.T.nsmallest(2, age_group).values.flatten().tolist()
    print(f"Age Group {age_group}:")
    for service, value in zip(lowest_two_services, lowest_two_values):
        print(f"  {service}: {round(value, 2)}")
```
